The objective is to create a class that takes a key and an alphabet and is able to encode and decode messages via the Vigen√®re cipher.
## My Solution

```python
class VigenereCipher(object):
    def __init__(self, key, alphabet):
        self.key=key
        self.alphabet=alphabet
        
    def as_number(self,text):
        output=[]
        for element in text:
            if element in self.alphabet:
                output.append(self.alphabet.index(element))
            else:
                output.append(element)
        return output
    
    def process_text(self,text,direction):
        output=""
        number_key=self.as_number(self.key)
        number_text=self.as_number(text)        

        for i in range(len(text)):
            if type(number_text[i]) is str: output+=text[i] # Ignore Chars that are not in the alphabet
            elif direction=="encode": 
                output+=(self.alphabet*10)[(number_key*10)[i]+(number_text*10)[i]]  
            elif direction=="decode":
                output+=(self.alphabet*10)[(number_text*10)[i]-(number_key*10)[i]]
                
        return output
                
    def encode(self, text):
        return self.process_text(text,"encode")
             
    def decode(self, text):
        return self.process_text(text,"decode")
```
